>> # ___Visualization___

```python
sns.set()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sales', bins=50, palette='plasma',edgecolor='black')
plt.title('Histogram of Sales',fontsize=18)
sns.set_style("whitegrid")
plt.xlabel('Sales',fontsize=14)
plt.ylabel('Frequency',fontsize=14)
plt.show()
```

___From the histogram we can understand that in the supermarket that most of the  sales are less than 5000 and the frequency of around 1000 is most.___

```python
x=df['Year']
y=df['Sales']

plt.ticklabel_format(style='plain')

sns.lineplot(x=x,y=y,marker='o',color='green')
plt.title('Total Sales over Years',fontsize=18)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(x.astype(int))
plt.grid(True)
plt.show()
```

___In this plot we could see the yearly sales.___

```python
sales_by_years = df.groupby('Year')['Sales'].sum().reset_index()
plt.figure(figsize=(8,8))
plt.pie(sales_by_years['Sales'],labels=sales_by_years['Year'],autopct='%1.1f%%',startangle=90,counterclock=False,explode=[0,0,0,0.1])
plt.title('Sales Distribution by Years',fontsize=18)
plt.show()
```

___From this plot it can be seen that 2018 year has the most sales.___

```python
plt.figure(figsize=(14,6))
sns.barplot(x='Sub-Category',y='Sales',data=df,palette='plasma',order=df['Sub-Category'].value_counts().head(100).index,errorbar=None)
plt.title('Sales by Sub-Categories',fontsize=18)
plt.xlabel('Sub-Category',fontsize=13)
plt.ylabel('Sales',fontsize=13)
plt.tick_params(axis='x',rotation=75)
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['number'])

correlation_matrix = numeric_columns.corr()
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(correlation_matrix,annot=True,cmap='plasma',linewidth=0.5)
plt.title('Correlation Matrix of Numeric Variables',fontsize=18)
plt.show()
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum().reset_index()
plt.figure(figsize=(8,8))
plt.pie(sales_by_years['Sales'],labels=sales_by_region['Region'],autopct='%1.1f%%',startangle=90,counterclock=False,explode=[0,0,0,0.1])
plt.title('Sales by Regions',fontsize=18)
plt.show()
```

___West region has the most sales.___

```python
selected_columns = ['Segment','Sales','Year']
selected_data = df[selected_columns]

segment_yearly_scales = selected_data.groupby(['Segment','Year'])['Sales'].sum().reset_index()

sns.barplot(x='Year',y='Sales',data=df,hue='Segment',palette='Dark2',errorbar=None)
plt.title('Sales Trends for Different Segments')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Segment',bbox_to_anchor=(1.05,1),loc='upper left')
plt.show()
```
